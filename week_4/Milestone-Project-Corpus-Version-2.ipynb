```python
#tokenize and lemmaztize
from nltk.tokenize import WhitespaceTokenizer

X_train_CLEAN = []

#tokenizer = TreebankWordTokenizer()
tokenizer = WhitespaceTokenizer()
lemmatizer = WordNetLemmatizer()


for doc in X_train["Lyrics"]:
    tokens = tokenizer.tokenize(text=doc)
    clean_doc = " ".join(lemmatizer.lemmatize(token) for token in tokens)
    X_train_CLEAN.append(clean_doc)
```

```python
#tokenize and lemmaztize
from nltk.tokenize import WhitespaceTokenizer

X_test_CLEAN = []

#tokenizer = TreebankWordTokenizer()
tokenizer = WhitespaceTokenizer()
lemmatizer = WordNetLemmatizer()

for doc in X_test["Lyrics"]:
    tokens = tokenizer.tokenize(text=doc)
    clean_doc = " ".join(lemmatizer.lemmatize(token) for token in tokens)
    X_test_CLEAN.append(clean_doc)
```
